```python
import numpy.linalg as LA
cmap=plt.cm.autumn
```

```python
b_start=200
test_m=data3.loc[b_start,['mx','my','mz']].values
test_m_norm=LA.norm(test_m)
result=[]
for item in data1.iterrows():
    thism=item[1][['mx','my','mz']].values
    result.append((test_m*thism).sum()/LA.norm(thism)/test_m_norm)
result=np.array(result)
```

## 小结

以上的部分都是对于相同设备的分析，目前可以得出以下结论：
+ 单纯使用强度值去匹配，需要匹配的线段比较长，一般来说应超过２０步，也就是大约10m
+ 如果考虑使用两个维度的数据，即【my,mz】能够有效缩短这个匹配长度，甚至可以低到只有3,5个点
+ 这都是使用dtw来匹配的
+ 对于同一个设备而言，甚至可以建立地磁三个维度的指纹库，【mx,my,mz】他至少是一个必要条件，接着使用粒子滤波是完全有希望变成全局定位的

下面要着力解决的问题是确认不同设备是否有这种特性，以及在匹配完成的时候的移动策略

```python
result=[]
for test_start in range(50,dataA.shape[0]-100,5):
    wright_index=processDtw(test_start,100,dataA,data1,'mmagn')
    for len_last in range(1,20):
        len_last*=5
        this_index=processDtw(test_start,len_last,dataA,data1,'mmagn')
        if abs(this_index-wright_index)<15:
            result.append([test_start,len_last,1])
        else:
            result.append([test_start,len_last,0])
```

## 单匹配点的整体计算

考虑到dtw过程中，是带有距离缩放的，虽然观测序列不带有缩放，但是候选的指纹序列是自然带有长短缩放的，此时如何将匹配正确的序列移动到候选序列上
从而求得移动量，或者当前坐标的量是非常关键的。
考虑到现在使用多点的匹配，且dtw正常情况下是按照序列的顺序进行匹配的，所以可以使用匹配上的单点进行移动，实际在其中具有伸缩和移动2钟情况
自然考虑最小二乘求解最优旋转矩阵
然后移动当前的点

```python
# %matplotlib inline
def show2d(b_start,startindex,lastlen,data_test,data_fp,colnum_name):
    distance,path=fastdtw(data_fp.loc[startindex:startindex+lastlen,colnum_name].values,data_test.loc[b_start:b_start+lastlen,colnum_name].values)
    plt.figure(figsize=(25,10))
    ax = plt.subplot(111)  # 创建一个三维的绘图工程
    #  将数据点分成三部分画，在颜色上有区分度
    ax.scatter(data_fp.loc[:,'x'], data_fp.loc[:,'y'], c='r',marker='*')  # 绘制数据点
    ax.scatter(data_test.loc[:,'x'], data_test.loc[:,'y'], c='gray',marker='s')  # 绘制数据点
    ax.scatter(data_test.loc[b_start:b_start+lastlen,'x'], data_test.loc[b_start:b_start+lastlen,'y'], c='g',s=50)  # 绘制数据点
    ax.scatter(data_fp.loc[startindex:startindex+lastlen,'x'], data_fp.loc[startindex:startindex+lastlen,'y'], c='b',marker='o',s=50)  # 绘制数据点

    for item in path:
        ax.plot([data_fp.loc[item[0]+startindex,'x'],data_test.loc[item[1]+b_start,'x']],[data_fp.loc[item[0]+startindex,'y'],data_test.loc[item[1]+b_start,'y']],'c')
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    plt.grid()
    plt.show()
    print (distance)
    return distance,path
```

## 小结

求解旋转矩阵，由于这里并不能完整的给出全部的匹配点，所以如果匹配点有问题的话，计算出的旋转矩阵是有问题的
特别是在比较长的时候，旋转矩阵容易带出一些误差，直接移动点反而更好看一下
这样的话，就要要求指纹必须足够多，起码覆盖所有的场合
同时，也要考虑的是，单单经过缩放，可以说一定会对后面的角度产生影响
有一个办法是，匹配长度不太长，这样误差也是可控的

从结果上看，直接把点移动过来可能反而是最快的方法，本来我的设想是这有助于后面去优化步长，
那现在看来，直接优化步长即可，坐标不动，后面直接优化长度。而使用花哨的那种旋转的问题是，他延迟不了太长时间
这意味着你可能得不断的用一小段去匹配，然后去旋转，这样对于过去的时间的有限的，但是对于当下来说，没有什么
太大的意义。
